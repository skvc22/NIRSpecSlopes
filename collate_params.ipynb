```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
from astropy.io import fits
```

```python
filename = 'wasp178b-nrs1.xc'
lc_dir = './lcs/'
```

```python
ds = xr.open_dataset(lc_dir+filename, engine='h5netcdf')
    
times = ds['time_flux'].values
flux = ds['flux'].values
wavelength = ds['wavelength'].values

print(times.shape)
print(flux.shape)
print(wavelength.shape)
```

```python
names=['hatp1b', 'hatp26b', 'toi134b', 'wasp63b', 'wasp107b', 'wasp178b']
vmag=[9.8, 11.8, 11, 11.2, 11.6, 9.9]
kmag=[8.9, 9.6, 7.1, 9.4, 8.6, 9.7]
jmag=[9.156, 10.080, 7.941, 9.818, 9.378, 9.775]
teff=[5980, 5079, 3842, 5550, 4425, 9360]
rp=[1.319, .63, 0.1454, 1.41, .94, 1.81] #jupiter radii
# integnum=[1336, 481, 3359, 1351, 1230, 605] #if it should have been the other axis its a constant 2048
ngroups=[]
ninteg=[]
expstart=[]
```

```python
for i in names:
    hdu = fits.open('./lcs/fits/'+i+'.fits')
    ngroups.append(hdu[0].header['NGROUPS']) #nints, expstart
    ninteg.append(hdu[0].header['NINTS']) 
    expstart.append(hdu[0].header['EXPSTART']) 
```

```python
dct = {'names':names, 'v_mag':vmag, 'j_mag':jmag, 'k_mag':kmag, 'Teff':teff, 'R_p':rp, 'groups':ngroups, 'integrations':ninteg, 'exposure_start':expstart}
table=pd.DataFrame(dct)
table
```

```python
table.sort_values(by='j_mag')
```

```python
table.to_csv('params.csv')
```
