```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds1 = xr.open_dataset('./outputs/toi134b-nrs1/fit.xc', engine='h5netcdf')
    
times1 = ds1['time_flux'].values
flux1 = ds1['raw_flux'].values
model1 = ds1['final_model'].values
resid1 = ds1['resid'].values

ds2 = xr.open_dataset('./outputs/toi134b-nrs2/fit.xc', engine='h5netcdf')
    
times2 = ds2['time_flux'].values
flux2 = ds2['raw_flux'].values
model2 = ds2['final_model'].values
resid2 = ds2['resid'].values
```

```python
fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
plt.subplots_adjust(hspace=0, wspace=.125)
fig.set_figheight(6)
fig.set_figwidth(12)

ax1.errorbar(times1, flux1, yerr=0, fmt='.', zorder=0, alpha=0.2, label='raw flux', color='gray')
ax1.plot(times1, model1, zorder=1, color='purple', label='transit + systematics')

ax1.set_ylabel('Relative Flux',fontsize=14)
ax1.legend(frameon=False,loc='lower left')
ax1.set_ylim(.998, 1.002)
ax1.set_xlim(times1.min()-0.001, times1.max()+0.001)

ax3.errorbar(times1, (resid1)*1e6, yerr=None, fmt='.',color='gray', alpha=0.2)        
ax3.set_xlabel('BJD (days)',fontsize=14)
ax3.set_ylabel('Residuals (ppm)',fontsize=14)
ax3.axhline(y=0,color='purple',linewidth=2, zorder=10)


ax2.errorbar(times2, flux2, yerr=0, fmt='.', zorder=0, alpha=0.2, label='raw flux', color='gray')
ax2.plot(times2, model2, zorder=1, color='purple', label='transit + systematics')

ax2.legend(frameon=False, loc='lower left')
ax2.set_ylim(.998, 1.002)
ax2.set_xlim(times2.min()-0.001, times2.max()+0.001)

ax4.errorbar(times2, (resid2)*1e6, yerr=None, fmt='.',color='gray', alpha=0.2)
           
ax4.set_xlabel('BJD (days)',fontsize=14)
ax4.axhline(y=0,color='purple',linewidth=2, zorder=10)

ax3.set_yticks([-500, -250, 0, 250, 500]);
ax4.set_yticks([-500, 0, 500, 1000, 1500]);

ax1.text(.015, .9,'NRS1', size='x-large', transform=ax1.transAxes)
ax2.text(.015, .9,'NRS2', size='x-large', transform=ax2.transAxes)
```
